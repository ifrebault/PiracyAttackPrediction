# Séparation en zones

```python
cellWidthNq = 35
cellWidthKm = cellWidthNq * 1.852
print(cellWidthKm)
```

```python
print(dfAden.index.values)
```

```python
zonesAden = []

for i in dfAden.index.values : 
    for j in range(0,70) :
        if (dfAden.longitude[i] + abs(min(dfAden.longitude)))/j<cellWidthKm :
            zonesAden.append(j)
        if (dfAden.latitude[i] + abs(min(dfAden.latitude)))/j<cellWidthKm :
            zonesAden.append(j)
print(zonesAden)
        
        
```
